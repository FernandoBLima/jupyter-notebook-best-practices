### Estrutura do diretório
A estrutura do projeto tende a ser flexível! 

Consistência é uma das coisas mais importantes.

```
	|- notebooks/
    |- data/
	|- src/
    |- reports/
	|- utils/
	|- environment.yml
	|- README.md
```

Exemplo de estrutura completa:

```
	|- notebooks/
	   |- 00-notebook.ipynb
	   |- 01-notebook.ipynb
	   |- archive/
		  |- no-longer-useful.ipynb
       |- figures/
          |- image1.png
	|- src/
        |- __init__.py
        |- config.py
        |- data.py
        |- utils.py
	|- README.md
    |- reports
       |- report1.csv
       |- report2.pdf
	|- data/
	   |- raw/
	   |- processed/
	   |- intermeadiate/
	|- utils/
	   |- script1.py
	   |- script2.py
	   |- archive/
	      |- no-longer-useful.py
	|- environment.yml
```

- Mas afinal, quando ser inconsistente!?
- Às vezes, as recomendações do guia de estilo simplesmente não são aplicáveis. 
- Em caso de dúvida, use seu bom senso. E não hesite em perguntar!

### Data Pipeline
   
Os conjuntos de dados devem ser imutáveis.
   
```
	|- data/
        |- raw/ 
        |- processed/ 
        |- intermeadiate/
        |- external/ 
```

- raw: Diretório de dados brutos, somente para a leitura
- processed: Dados obtidos após a realização do pré-processamento e limpeza.
- intermeadiate:  Formato intermediário entre bruto e processado: Models, Outputs, Reports
- external: Dados considerados externos como: dicionários, sinônimos, etc.

### Como organizar um notebook

- Notebooks não devem ser longos, senão seria chamado Jupyter Book :)
    - Quebrar em arquivos lógicos menores 
    - Criar notebooks específicos com uma ordenação de arquivos
- Recomendável mover o código para pacotes autônomos
- Adicionar títulos e descrições em Markdown para explicar os passos
    - Documentar permitirá um maior entendimento, assim como fornecer orientação adicional para modificações
    - Documentar não é comentar o código

- As células devem conter apenas uma saída lógica e possuir uma order
- Tratar erros na saída das células
- Importar código ao ínves de notebook
- Criar arquivos com 'logical chuncks'
    - {algo-lógico}-notebook.ipynb 

Organizando os notebooks:

```
	|- notebooks/
	   |- 00-preprocessing.ipynb
	   |- 01-transformation.ipynb
       |- 02-modeling.ipynb
       |- 03-evaluation.ipynb
       |- 04-reports.ipynb
       |- utils/
          |- plots.py
          |- preprocessing.py
	   |- archive/
		  |- no-longer-useful.ipynb
       |- figures/
          |- image1.png
       
```

````python
import pandas as pd

# Não faça isso
mnth = pd.DataFrame(data = {'janeiro': [1, 2], 'fevereiro': [3, 4], 'março': [5, 6], 'abril': [7, 8]})
mnth = mnth.filter(items=['janeiro', 'fevereiro'])

# Faça isso
month_df = pd.DataFrame(data = {'janeiro': [1, 2], 'fevereiro': [3, None], 'março': [5, 6], 'abril': [7, None]})
monthly_filter_df = month_df.filter(items=['janeiro', 'fevereiro'])
````

##### Don’t repeat yourself - DRY 
##### Automatize tarefas repetitivas
##### Mantenha seu código organizado em módulos.

````python
monthly_filter_df.fillna({'fevereiro': 0})
monthly_filter_df.groupby(['fevereiro']).mean().sort_values(by='fevereiro')

target = 'fevereiro'
````

##### Uma célula deve ter uma saída e finalidade
##### As funções devem conter um objetivo. 
##### A divisão de funções leva à reutilização e testes mais fáceis.
##### Não use flags como parâmetros de função

````python
def split_email(email: str, temp: bool):
    if temp:
        return email.split('@')[0]
    else:
        return email.split('@')[1]

split_email('ola@tudobom.com', True)
````

## Como aplicar os famosos teste unitários?

Plano para erros a.k.a Programação defensiva.  Escreva e execute testes.

- asserts
- Doctest
- unittest

### Asserts

````python
def multiplyBy(x, y):
    return x * y

result = multiplyBy(2,3)
assert result == 6
````

### doctest

````python
def add(a, b):
    '''
    This is a test:
    >>> add(2, 3)
    5
    '''
    return a + b
````

````python
import doctest
doctest.testmod(verbose=True)
````

### unittest

````python
import unittest

class TestNotebook(unittest.TestCase):
    
    def test_add(self):
        self.assertEqual(multiplyBy(5,2), 2)
        
unittest.main(argv=[''], verbosity=2, exit=False)
````
